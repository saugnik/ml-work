```python
import numpy as np
import seaborn as sns
from tensorflow.keras.utils import load_img,img_to_array
import matplotlib.pyplot as plt
import os
```

```python
pic_size =2160


plt.figure(0, figsize=(10,20))
cpt = 0

for expression in os.listdir(data ):
    for i in range(1,6):
        cpt = cpt + 1
        plt.subplot(7,5,cpt)
        img = load_img(data + "/"+expression + "/" +os.listdir(data+"/"+ expression)[i], target_size=(pic_size, pic_size))
        plt.imshow(img, cmap="gray")

plt.tight_layout()
plt.show()
```
